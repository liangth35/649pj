```python
import numpy as np
import pandas as pd
import altair as alt
from vega_datasets import data
df=pd.read_csv('pj2.csv')
df
```

```python
question="Very confident about providing same quality of care for PWD"
Characteristics=["Gender","Race/ethnicity","Primary specialty","Practice type"]
def getMosaic(df, question, Characteristic):    
    df1=df[df['Characteristic'] == Characteristic]
    base = alt.Chart(df1
    ).transform_stack(
        stack="Percent",
        as_=["x1", "x2"],
        groupby=['Answer'],
    ).transform_stack(
        stack=question,
        as_=["y1", "y2"],
        groupby=['Type'],
    ).transform_calculate(
        xc="(datum.x1+datum.x2)/2",
        yc="(datum.y1+datum.y2)/2",
    )

    rect = base.mark_rect().encode(
        x=alt.X("x1:Q", axis=None),
        x2="x2",
        y=alt.Y("y1:Q", axis=None),
        y2="y2",
        color=alt.Color("Type:N", legend=None),
        opacity=alt.Opacity("Answer:N", legend=None),
        tooltip=[alt.Tooltip("Answer:N",title=f"{question}?"), alt.Tooltip(f"{question}:Q", format='.1%', title="percentage")],
    )

    text = base.mark_text(baseline="middle").transform_filter(alt.datum.Answer == "Yes").encode(
        x=alt.X("xc:Q", axis=None), y=alt.Y("yc:Q"), text=alt.Text(f"{question}:Q",format='.1%')
    )

    labels = base.mark_text(baseline="middle", align="center").encode(
        x=alt.X(
            "xc:Q",
            axis=alt.Axis(title=None, orient="top"),
        ),
        color=alt.Color("Type", legend=None),
        text=alt.Text("Type:N"),
    )
    mosaic = rect + text
    res = (labels & mosaic).properties(title=alt.TitleParams(text=Characteristic, anchor="middle"))
    return res


[mosaic1,mosaic2,mosaic3,mosaic4] = [getMosaic(df, question, Characteristics[i]) for i in range(len(Characteristics))]

fig = (mosaic1|mosaic2|mosaic3|mosaic4
).configure_view(stroke=""
).configure_concat(spacing=10
).configure_axis(domain=False, ticks=False, labels=False, grid=False)


fig.title=alt.TitleParams(text='The percentage of physicians of different charactistics who are "very confident" about providing same quality of care for disabled people', anchor="middle")
fig
```
